# 04. Diffusion-Reaction CNO Starter

This notebook provides a simple convolutional operator baseline in the spirit of a CNO.

It is not a full research-grade CNO implementation. The goal is to give your repository an approachable, extensible convolution-based operator example beside DeepONet and FNO.


```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
```

```python
rng = np.random.default_rng(11)
torch.manual_seed(11)

n_samples = 256
n_grid = 128
x = np.linspace(0, 1, n_grid)

inputs = []
targets = []
for _ in range(n_samples):
    centers = rng.uniform(0.2, 0.8, size=2)
    widths = rng.uniform(0.05, 0.12, size=2)
    forcing = sum(np.exp(-((x - c) ** 2) / (2 * w ** 2)) for c, w in zip(centers, widths))
    kernel = np.exp(-40 * (x - 0.5) ** 2)
    kernel /= kernel.sum()
    smoothed = np.convolve(forcing, kernel, mode="same")
    target = 0.8 * smoothed + 0.2 * np.tanh(forcing)
    inputs.append(forcing)
    targets.append(target)

X = torch.tensor(np.array(inputs), dtype=torch.float32).unsqueeze(1)
Y = torch.tensor(np.array(targets), dtype=torch.float32).unsqueeze(1)
loader = DataLoader(TensorDataset(X, Y), batch_size=16, shuffle=True)
```

```python
class TinyCNO1d(nn.Module):
    def __init__(self, width=32):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv1d(1, width, kernel_size=5, padding=2),
            nn.GELU(),
            nn.Conv1d(width, width, kernel_size=5, padding=2),
            nn.GELU(),
            nn.Conv1d(width, width, kernel_size=5, padding=2),
            nn.GELU(),
            nn.Conv1d(width, 1, kernel_size=1),
        )

    def forward(self, x):
        return self.net(x)


model = TinyCNO1d()
optimizer = torch.optim.Adam(model.parameters(), lr=1e-3)
loss_fn = nn.MSELoss()
```

```python
for epoch in range(30):
    running = 0.0
    for xb, yb in loader:
        optimizer.zero_grad()
        pred = model(xb)
        loss = loss_fn(pred, yb)
        loss.backward()
        optimizer.step()
        running += loss.item()
    if (epoch + 1) % 10 == 0:
        print(f"epoch {epoch + 1:02d} | loss {running / len(loader):.6f}")
```

```python
with torch.no_grad():
    pred = model(X[:1]).squeeze().numpy()
truth = Y[0].squeeze().numpy()
source = X[0].squeeze().numpy()

plt.figure(figsize=(9, 4))
plt.plot(x, source, label="input field")
plt.plot(x, truth, label="target")
plt.plot(x, pred, "--", label="Tiny CNO prediction")
plt.title("Convolutional operator baseline")
plt.legend()
plt.tight_layout()
plt.show()
```

## Takeaway

This notebook gives you a clean convolutional-operator entry point for GitHub visibility without over-claiming a full CNO reproduction.
