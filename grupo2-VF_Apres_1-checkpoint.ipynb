```python
#importar as bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

#pairplot
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable

#Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import GridSearchCV, KFold, cross_validate
from sklearn.metrics import (mean_absolute_error, mean_squared_error,
                             explained_variance_score, r2_score)


from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, HistGradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
from sklearn.svm import SVR
import warnings
#Feature importance
from sklearn.inspection import permutation_importance

warnings.filterwarnings('ignore')# Suppress convergence warnings for cleaners -> apenas utilizar depois de ler todos os warnings originais

# Selecionar uma random seed para dar reprodutibilidade
np.random_seed = 42
random_state = 42

# importar data
df = pd.read_csv('training_data_set.csv')
print(df.shape)
train_df, test_df = train_test_split(
    df,
    test_size = 0.2,
    random_state = 42 # para fixar os dados que são gerados na primeira "corrida"
)

#Verificar informação sobre os dados
print("Train size:", train_df.shape)
print("Test size:", test_df.shape)

print("Train type:\n",train_df.dtypes)
print("Train isnull:\n",train_df.isnull().sum())
print("Train unique:\n",train_df.nunique())
print("Train describe:\n",train_df.describe().T)

#eliminar variáveis que não são para testar (treino e teste)
train_df = train_df.drop(['strainRate_p50', 'k_p50','Umag_p50'], axis=1) 
test_df = test_df.drop(columns=['strainRate_p50','k_p50','Umag_p50'] )

#criar o histograma
train_df.hist(bins=30, figsize=(12, 8))
plt.tight_layout()
plt.show()
```

<a name="6-execução-da-pipeline--implementação-do-professor"></a>
## 6. Execução da Pipeline — Implementação do Professor
- Separar colunas por tipo: float_cols, int_cols.
- Construção da pipeline: Pipeline([('prep', preprocess), ('model', estimator)]).
- Definição de múltiplos modelos e respetivos grids de hiperparâmetros.
- Aplicação de GridSearchCV com validação cruzada (KFold com 5 folds).
- Avaliação de métricas nos dados de treino e no conjunto de teste final.
- Armazenamento e comparação do desempenho dos modelos.
- Seleção do melhor modelo com base no menor CV-RMSE.
- Visualização dos resultados: comparação entre valores previstos e reais (gráficos scatter com métricas).

At this point you should have the following Pandas DataFrames/Series pre-processed:
- **y_train** (the output values for the train data)
    - shape: (n_rows_train, 1) if it is a Pandas **DataFrame**  
    - shape: (n_rows_train, ) if it is a Pandas **Series**
- **y_test**  (the output values for the test data)
    - shape: (n_rows_test, 1) if it is a Pandas **DataFrame** 
    - shape: (n_rows_test, ) if it is a Pandas **Series**
- **X_train** (the features of the train data - **not scaled** yet)
    - shape:  (n_rows_train, n_features) - should be a Pandas **DataFrame**
- **X_test** (the features of the test data - **not scaled** yet) 
    - shape:  (n_rows_test, n_features) - should be a Pandas **DataFrame**


```python
# if y_train and y_test are a single column DataFrame converts to a Series
y_train = y_train.squeeze()
y_test = y_test.squeeze()
```

```python
# if you have other types of features, add as necessary
float_cols = X_train.select_dtypes(include='float64').columns.tolist()
int_cols   = X_train.select_dtypes(include='int64').columns.tolist()

print(f"float64 type columns: {float_cols}\n")
print(f"int64 type columns: {int_cols}")
```

```python
# ------------------------------------------------------------------
# 1. Pre-processing block: scaling of numeric (float, int) columns
#    IMP.: If you have other data types, adjust as necessary 
#          (currently is dropping all other type features)
# ------------------------------------------------------------------
preprocess = ColumnTransformer(
    transformers=[("num", StandardScaler(), float_cols),
                  ("int", MinMaxScaler(), int_cols) 
                 ],
    remainder="drop"   # drop any non-numeric cols
)
```

```python
# ------------------------------------------------------------------
# 2. Models and their hyper-parameter grids
#    You need to select the adequate ranges of hyperparameters 
#    that are more relevant - note that if you select too many
#    hyperparameters to test several values, the fitting and 
#    optimizations will be very time-consuming
# ------------------------------------------------------------------

models = {
    'Linear Regression': (
        LinearRegression(),
        {}
    ),

    'Linear Regression Reg L1: Lasso': (
        Lasso(fit_intercept = True, random_state = random_state),
        {
            'model__alpha': [0.05, 0.1, 0.2],
        }
    ),

    'Linear Regression Reg L2: Ridge': (
        Ridge(fit_intercept = True, random_state = random_state),
        {
            'model__alpha': [0.05, 0.1, 0.2],
        }
    ),

    'Decision Tree': (
        DecisionTreeRegressor(random_state = random_state),
        {
            'model__max_depth': [3, 5, 10, 15, None],
            'model__min_samples_split': [2, 5, 10],
            'model__min_samples_leaf': [1, 2, 4],
            'model__max_features': ['sqrt', 0.5, None]
        }
    ),
    
    'Random Forest': (
        RandomForestRegressor(random_state = random_state),
        {
            'model__n_estimators': [200, 500],
            'model__max_depth': [None, 5, 10, 15],
            'model__min_samples_split': [2, 5, 10],
            'model__max_features': ['sqrt', 'log2', 0.5]
        }
    ),
    
    'Gradient Boosting (sklearn)': (
        HistGradientBoostingRegressor(early_stopping=True, tol = 1.0e-7, \
                            validation_fraction = 20, n_iter_no_change = 50, 
                            random_state = random_state),
        {
            'model__learning_rate': [0.1, 0.2, 0.5],
            'model__max_iter': [200, 500, 1000],
            'model__max_depth': [3, 4, 5],
            'model__l2_regularization': [0.0, 0.1, 0.2]
        }
    ),
    
    'KNN': (
        KNeighborsRegressor(),
        {
            'model__n_neighbors': [3, 5, 11, 15, 21],
            'model__weights': ['uniform', 'distance'],
            'model__p': [1, 2],
            'model__leaf_size': [10, 20, 50]
        }
    ),
    
    'SVR': (
        SVR(),
        {
            'model__kernel': ['rbf', 'linear'],
            'model__C': [0.1, 0.5], 
            'model__gamma': ['scale', 'auto'],
            'model__epsilon': [0.1]              
        }
    ),

    'FCNN': (
        MLPRegressor(random_state = random_state, max_iter = 2000, \
                     early_stopping = True, validation_fraction = 0.2, \
                     n_iter_no_change = 50, tol = 1e-4),
        {
            'model__hidden_layer_sizes': [ (20, 20), (50, 50), (20, 50, 50, 20)],
            'model__activation': ['relu', 'tanh'],
            'model__learning_rate_init': [0.001]
        }
    )

}
```

```python
# ------------------------------------------------------------------
# 3. Loop over models, tune, cross-validate and evaluate on test set
# ------------------------------------------------------------------

n_splits = 5      # number of folds used in CV

cv = KFold(n_splits=n_splits, shuffle=True, random_state = random_state)
scorers = {
    "RMSE": "neg_root_mean_squared_error",
    "MSE": "neg_mean_squared_error",
    "MAE": "neg_mean_absolute_error",
    "Explained Variance": "explained_variance",
    "R2": "r2"
}

cv_results, test_results = {}, {}

print(f"Hyperparameter tunning via grid search and using K-fold cross validation with {n_splits} folds")

for name, (estimator, grid) in models.items():
    print(f"\n*** {name} model ***\n    ", end = '')

    pipe = Pipeline(steps=[("prep", preprocess), ("model", estimator)])

    search = GridSearchCV(
        pipe,
        param_grid=grid,
        scoring="neg_root_mean_squared_error",
        cv=cv,
        n_jobs=-1,
        verbose = 2,
    )
    search.fit(X_train, y_train)
    
    if len(search.best_params_) > 0:
        print("    Best hyper-parameters found:")
        for key, value in search.best_params_.items():
            print(f'      + {key.replace("model__", "")}: {value}')

    print(f'    Metrics computed using CV:')
    # Cross-validation metrics for the best hyperparameters
    cv_metrics = cross_validate(
        search.best_estimator_,
        X_train,
        y_train,
        cv=cv,
        scoring=scorers,
        n_jobs= -1,
        return_train_score = True,
        verbose = 0
    )

    cv_metrics = pd.DataFrame(cv_metrics)

    neg_metrics = {"RMSE", "MSE", "MAE"}        # those that came from "neg_*" scorers

    def pos(arr, metric):        # flip sign for the neg_metrics (minimize)
        return -arr if metric in neg_metrics else arr

    summary = {
        m: (pos(cv_metrics[f"test_{m}"], m).mean(),
            pos(cv_metrics[f"test_{m}"], m).std())
        for m in ["RMSE", "MSE", "MAE", "Explained Variance", "R2"]
    }
    
    cv_results[name] = summary
    
    for m, (mean, std) in summary.items():
        metric_name = (m + 20*'.')[0:20]
        print(f"      {metric_name:20s}: {mean:7.4f} ± {std:7.4f}")

    # ── Test-set evaluation ────────────────────────────────────────
    y_pred = search.best_estimator_.predict(X_test)

    test_results[name] = {
    "RMSE": np.sqrt(mean_squared_error(y_test, y_pred)),
    "MSE":  mean_squared_error(y_test, y_pred),
    "MAE":  mean_absolute_error(y_test, y_pred),
    "Explained Variance": explained_variance_score(y_test, y_pred),
    "R2":   r2_score(y_test, y_pred),
    "y_pred": y_pred,
    }

# ------------------------------------------------------------------
# 4. Pick the overall best algorithm (lowest mean CV-RMSE)
# ------------------------------------------------------------------

best_model_name = min(cv_results, key=lambda k: cv_results[k]["RMSE"][0])

# ------------------------------------------------------------------
# 5. Print the metrics
# ------------------------------------------------------------------

print(f"\n{65*'='}\n\nSelected best model by CV-RMSE: ***{best_model_name}***")

print(f"\n{65*'='}\n\nMetrics computed on the unseen test-set:")
for name, metrics in test_results.items():
    print(f"\n{name}:")
    for m in ["RMSE", "MSE", "MAE", "Explained Variance", "R2"]:
        metric_name = (m + 20*'.')[0:20]
        print(f"  {metric_name:20s}: {metrics[m]:7.4f}")
```

> **Note:** In a real-world deployment (where we only have features, not true *y*), we wouldn’t generate all the following plots for unseen data, because we don’t know the true outputs.   
> Instead, we should apply only the single best model to new inputs and deploy its predictions.  
> Here we are comparing all models (with their best hyperparameters determined via grid search) with the *known* test data, just to show the predictive capabilities of each model on previously unseen data.

```python
# ------------------------------------------------------------------
# 6. Scatter plots: ŷ vs y  (test set)
# ------------------------------------------------------------------
print(f"\n{70*'='}\n{10*' '}Predictions vs. True values on the unseen Test-set:\n{70*'='}\n")

n_models = len(test_results)
n_cols   = 2
n_rows   = (n_models+n_cols-1) // n_cols

fig, axes = plt.subplots(
    n_rows, n_cols,
    figsize=(4 * n_cols, 4 * n_rows),
    squeeze=False         # always 2-D array
)

for ax, (name, res) in zip(axes.ravel(), test_results.items()):
    # scatter diagonal reference line
    ax.scatter(y_test, res["y_pred"], alpha=0.6)
    lims = [min(y_test.min(), res["y_pred"].min()),
            max(y_test.max(), res["y_pred"].max())]
    ax.plot(lims, lims, linestyle="--")

    # metrics panel
    metrics_text = (
        f"RMSE.....: {res['RMSE']:7.4f}\n"
        f"MSE......: {res['MSE']:7.4f}\n"
        f"MAE......: {res['MAE']:7.4f}\n"
        f"Expl.Var.: {res['Explained Variance']:7.4f}\n"
        f"R².......: {res['R2']:7.4f}"
    )
    ax.text(
        0.05, 0.95, metrics_text,
        transform=ax.transAxes,
        fontsize=10, fontfamily="monospace",
        verticalalignment="top",
        bbox=dict(boxstyle="round,pad=0.5",
                  facecolor="white", alpha=0.8)
    )

    # labels & layout
    ax.set(
        title=f"{name}: $\hat{{y}}$ vs $y$ (test)",
        xlabel="True y",
        ylabel="Predicted y"
    )

# Hide any unused subplot slots (e.g. when #models is odd)
for ax in axes.ravel()[len(test_results):]:
    ax.set_visible(False)

plt.tight_layout()
plt.show()
```

```python
# %%
# ------------------------------------------------------------------
# 7. Feature Importance via Permutation (Model-Agnostic)
# ------------------------------------------------------------------

# Correr o teste de importaância das features (Run permutation importance on the test set)
result = permutation_importance(
    estimator=search.best_estimator_,
    X=X_test,
    y=y_test,
    n_repeats=10,
    scoring='neg_root_mean_squared_error',
    random_state=42,
    n_jobs=-1
)

# Crear o dataframe com a informação anterior
perm_df = pd.DataFrame({
    "Feature": X_test.columns,
    "Importance (mean)": result.importances_mean,
    "Importance (std)": result.importances_std
}).sort_values(by="Importance (mean)", ascending=False)

# Plot dos resultados
plt.figure(figsize=(10, 6))
sns.barplot(x="Importance (mean)", y="Feature", data=perm_df)
plt.title("Permutation Feature Importance (Test Set)")
plt.tight_layout()
plt.show()
```

## 7. Resultados e conclusões




- O modelo que obteve melhor desempenho global foi o modelo FCNN (Fully Connected Neural Network) gerado pelo nosso código apresentou o menor erro de previsão com RMSE = 0.0198 e R² = 0.9616 para a variável epsilon_p50. No estudo de base (Random Forest para epsilon_p50), o RMSE é 0.0622 e R² = 0.6995 — ou seja, o nosso melhor modelo reduziu o erro em mais de 60% e aumentou significativamente o poder preditivo.
- A previsão de epsilon_p50, que está relacionada com dinâmica turbulenta e hidrodinâmica interna dos reatores, envolve interações complexas entre variáveis geométricas (ex: altura do líquido, número de pás, diâmetro do impeller) e operacionais (ex: rpm, volume). Essas relações dificilmente são lineares ou segmentadas. A FCNN é especialista em capturar relações altamente não-lineares, enquanto o Random Forest modela relações de forma piecewise (em blocos).
- A escolha de modelos variados e ajuste de hiperparâmetros foi decisiva: Testámos 9 algoritmos com GridSearchCV, incluindo redes neurais, árvores e métodos lineares. Isso permitiu-nos explorar diferentes tipos de técnicas e encontrar o modelo mais adaptado aos dados. O estudo de base limitou-se a Random Forest, sem detalhar tuning.
- O modelo reporta melhores resultados quando não são removidas as variáveis com correção entre 0.9 e 1, devido à baixa quantidade de variáveis no dataset original.
- Inicialmente, tentamos utilizar transformações log para as variáveis "volume", "liq_height" e "dia_imp_2", dado que o histograma identificava uma grande densidade de pontos com valor baixo.
